```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_centroid(data):
    centroid = np.mean(data, axis=0)
    return centroid

def calculate_distances(data, centroid):
    distances = np.linalg.norm(data - centroid, axis=1)
    return distances

def find_enclosing_distance(data, centroid, percentile):
    distances = calculate_distances(data, centroid)
    enclosing_distance = np.percentile(distances, percentile)
    return enclosing_distance

def plot_data_with_circle(data, centroid, enclosing_distance):
    distances = calculate_distances(data, centroid)
    enclosed_indices = np.where(distances <= enclosing_distance)[0]
    outside_indices = np.where(distances > enclosing_distance)[0]

    plt.scatter(data[outside_indices, 0], data[outside_indices, 1], color='red', label='Outside Circle')
    plt.scatter(data[enclosed_indices, 0], data[enclosed_indices, 1], color='blue', label='Inside Circle')
    plt.scatter(centroid[0], centroid[1], color='green', marker='x', s=100, label='Centroid')
    circle = plt.Circle((centroid[0], centroid[1]), enclosing_distance, color='orange', fill=False)
    plt.gca().add_patch(circle)
    plt.title('Data Points with Enclosing Circle')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.grid(True)
    plt.show()

# Generate a random dataset
np.random.seed(0)
num_points = 100
data = np.random.randn(num_points, 2)

# Calculate the centroid
centroid = calculate_centroid(data)

# Find the enclosing distance where 90% of the points fall
enclosing_distance = find_enclosing_distance(data, centroid, 90)

# Plot the data points with the enclosing circle
plot_data_with_circle(data, centroid, enclosing_distance)

```

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_centroid(data):
    centroid = np.mean(data, axis=0)
    return centroid

def calculate_distances(data, centroid):
    distances = np.linalg.norm(data - centroid, axis=1)
    return distances

def find_enclosing_distance(data, centroid, percentile):
    distances = calculate_distances(data, centroid)
    enclosing_distance = np.percentile(distances, percentile)
    return enclosing_distance

def plot_data_with_circle(data, centroid, enclosing_distance):
    distances = calculate_distances(data, centroid)
    enclosed_indices = np.where(distances <= enclosing_distance)[0]
    outside_indices = np.where(distances > enclosing_distance)[0]

    plt.scatter(data[outside_indices, 0], data[outside_indices, 1], color='red', label='Outside Circle')
    plt.scatter(data[enclosed_indices, 0], data[enclosed_indices, 1], color='blue', label='Inside Circle')
    plt.scatter(centroid[0], centroid[1], color='green', marker='x', s=100, label='Centroid')
    circle = plt.Circle((centroid[0], centroid[1]), enclosing_distance, color='orange', fill=False)
    plt.gca().add_patch(circle)
    plt.title('Data Points with Enclosing Circle')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.grid(True)
    plt.show()

def plot_resultant_data(resultant_data):
    plt.scatter(resultant_data[:, 0], resultant_data[:, 1], color='blue', label='Resultant Data')
    plt.title('Resultant Data Points')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.grid(True)
    plt.show()

# Generate a random dataset
np.random.seed(0)
num_points = 100
data = np.random.randn(num_points, 2)

# Calculate the centroid
centroid = calculate_centroid(data)

# Find the enclosing distance where 90% of the points fall
enclosing_distance = find_enclosing_distance(data, centroid, 90)

# Plot the data points with the enclosing circle
plot_data_with_circle(data, centroid, enclosing_distance)

# Filter out data points outside the enclosing circle
distances = calculate_distances(data, centroid)
enclosed_indices = np.where(distances <= enclosing_distance)[0]
resultant_data = data[enclosed_indices]

# Plot the resultant data points
plot_resultant_data(resultant_data)
```
